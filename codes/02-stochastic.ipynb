```python
import pandas as pd
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",header=None)
```

```python
import matplotlib
import matplotlib.pyplot as plt
y = df.iloc[0:100, 4].values
y = np.where(y == "Iris-setosa",-1,1)
X = df.iloc[0:100, [0,2]].values
plt.scatter(X[:50,0],X[:50,1], color="red", marker="o", label="setosa")
plt.scatter(X[50:100,0],X[50:100,1],color="blue",marker="x",label="versicolor")
plt.xlabel("sepal lenght [cm]")
plt.ylabel("petal lenght [cm]")
plt.legend(loc="upper left")
plt.show()
```

```python
from funzioni import *
```

```python
ada_gd=AdalineGD(n_iter=30,eta=1e-4,random_state=1)
ada_gd.fit(X,y)
plot_decision_regions(X,y,classifier=ada_gd)
plt.title("Adaline - Gradient Descent")
plt.xlabel('sepal length')
plt.ylabel('petal lenght')
plt.tight_layout()
plt.show()
plt.plot(range(1,len(ada_gd.cost_)+1),ada_gd.cost_,marker='o')
plt.tight_layout()
plt.show()
```

```python
ada_sgd=AdalineSGD(n_iter=15,eta=1e-2,random_state=1)
ada_sgd.fit(X,y)
plot_decision_regions(X,y,classifier=ada_sgd)
plt.title("Adaline - Stochastic Gradient Descent")
plt.xlabel('sepal length')
plt.ylabel('petal lenght')
plt.tight_layout()
plt.show()
plt.plot(range(1,len(ada_sgd.cost_)+1),ada_sgd.cost_,marker='o')
plt.tight_layout()
plt.show()
```

```python
X_std = np.copy(X)
X_std[:,0] = (X[:,0]-X[:,0].mean()) / X[:,0].std()
X_std[:,1] = (X[:,1]-X[:,1].mean()) / X[:,1].std()
```

```python
from classificatori import AdalineGD
```

```python
ada_gd = AdalineGD(n_iter=15,eta=0.01,random_state=1)
ada_gd.fit(X_std,y)
plot_decision_regions(X_std,y,classifier=ada_gd)
plt.tight_layout()
plt.show()

plt.plot(range(1,len(ada_gd.cost_)+1),ada_gd.cost_,marker='o')
plt.tight_layout()
plt.show()
```

```python
ada_sgd = AdalineSGD(n_iter=15,eta=0.01,random_state=1)
ada_sgd.fit(X_std,y)
plot_decision_regions(X_std,y,classifier=ada_sgd)
plt.tight_layout()
plt.show()

plt.plot(range(1,len(ada_sgd.cost_)+1),ada_sgd.cost_,marker='o')
plt.tight_layout()
plt.show()
```
